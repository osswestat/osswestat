### Cleaning URLs

```python
#Use the following regex pattern:

# \b(([\w-]+://?|www[.])[^\s()<>]+(?:\([\w\d]+\)|([^[:punct:]\s]|/)))


#df["links"]=df["project_urls"].astype(str)
pattern = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"
df_links = []
df_links = df["project_urls"].str.extract(pattern, expand=True)
```

```python
df_urls= pd.DataFrame(df_links)
```

```python
df_urls.head()
```

```python
df['links'] = df_urls[0]
```

```python
df['links'].sample(20)
```

```python
df.head()
```

```python
df_push = df.drop(['Metadata', 'python_versions', 'dependencies', 'project_urls', 'versions'], axis=1)
```

```python
df_push = df_push.sort_values(by=['PackageName'])
```
