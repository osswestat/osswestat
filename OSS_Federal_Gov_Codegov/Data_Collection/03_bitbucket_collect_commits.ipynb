```python
import requests
from requests.auth import HTTPBasicAuth
import re
import pandas as pd

# Your Bitbucket username and app password
username = 
app_password = 

# List of Bitbucket repository links
repo_links = [
    'https://bitbucket.org/berkeleylab/esdr-liblib/src/master/',
    # Add more repository links here
]

# Function to extract repository owner and name from URL
def extract_repo_info(url):
    pattern = r'bitbucket\.org/([^/]+)/([^/]+)/'
    match = re.search(pattern, url)
    if match:
        owner = match.group(1)
        repo = match.group(2)
        return owner, repo
    else:
        return None, None

# Function to get commit statistics for each repository
def get_commit_stats(owner, repo):
    commits_url = f'https://api.bitbucket.org/2.0/repositories/{owner}/{repo}/commits'
    response = requests.get(commits_url, auth=HTTPBasicAuth(username, app_password))

    if response.status_code == 200:
        commits = response.json().get('values', [])
        commit_stats = []
        for commit in commits:
            commit_hash = commit['hash']
            commit_details_url = f'https://api.bitbucket.org/2.0/repositories/{owner}/{repo}/commit/{commit_hash}'
            commit_details_response = requests.get(commit_details_url, auth=HTTPBasicAuth(username, app_password))
            if commit_details_response.status_code == 200:
                commit_details = commit_details_response.json()
                stats = {
                    'author': commit_details['author']['raw'],
                    'authorTimestamp': commit_details['date'],
                    'linesOfCode': {
                        'added': commit_details['stats']['lines_added'],
                        'deleted': commit_details['stats']['lines_removed']
                    },
                commit_stats.append(stats)
        return commit_stats
    else:
        print(f'Failed to fetch data for repository {repo}. HTTP Status code: {response.status_code}')
        return []

# Initialize a list to store all commit stats
all_commit_stats = []

# Fetching commit statistics for each repository
for link in repo_links:
    owner, repo = extract_repo_info(link)
    if owner and repo:
        commit_stats = get_commit_stats(owner, repo)
        all_commit_stats.extend(commit_stats)
    else:
        print(f'Failed to extract repository information from URL: {link}')

# Create a DataFrame from the commit statistics
df = pd.DataFrame(all_commit_stats)

print(df)
```
